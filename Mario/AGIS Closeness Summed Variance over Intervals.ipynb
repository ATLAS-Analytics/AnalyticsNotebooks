```python
%matplotlib inline
import matplotlib
matplotlib.rc('font', **{'size': 22})
```

```python
import json
```

```python
j = json.loads(open('totals_per_winsize.json').read())
stds = [0]*(len(j.keys())+1)
cars = [0]*(len(j.keys())+1)
for i in j:                                             
    stds[int(i.split('-')[1])] = j[i][1]
    cars[int(i.split('-')[1])] = j[i][0]
```

```python
import matplotlib.pyplot as plt
plt.plot(stds, 'g', label='Summed Variance')
plt.plot(cars, 'b', label='Summed Standard deviation')
plt.legend()
plt.title('AGIS Closeness Summed Variance over Intervals')
plt.xlabel('Aggregation interval in hours')
plt.ylabel('Value')
fig = plt.gcf()
fig.set_size_inches(17, 8)
```
