Program to show 3d geo map of CERN Point 1

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator
import numpy as np


from mpl_toolkits.mplot3d import Axes3D

fn='24931122'

with open(fn+'.asc', 'r') as f:
    lines=f.readlines()

print("lines:",len(lines))

dx, dy = 0.5, 0.5

# z = np.mgrid[slice(0,1000,dx), slice(0,1000.0,dy)]
z = np.zeros((2000,2000))
head=lines[:5]
lines=lines[6:]

for x in range(2000):
    l = lines[x]
    ys = l.split()
    for y in range(2000):
        z[x,1999-y]=float(ys[y])

x = np.arange(2000)
y = np.arange(2000)

cmap='Greys_r'


plt.figure(figsize=[10, 10])
plt.pcolormesh(x, y, z, cmap=cmap)#, norm=norm)
# im = ax0.contourf(x, y, z, levels=levels, cmap=cmap)
#fig.colorbar(im, ax=ax0)
#plt.colorbar()
plt.axis('off')
plt.savefig(fn, bbox_inches='tight')
plt.show()
```
