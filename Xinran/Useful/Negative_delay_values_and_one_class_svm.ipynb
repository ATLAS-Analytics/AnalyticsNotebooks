```python
time_window_width = 1      # hours
tww = time_window_width * 60 * 60 * 1000      # ms

count = 0
table_count = pd.DataFrame()
time_min = 1451607300000
time_max = 1464731700000

# Bin the original data into time_window_width, and count how many records are involved
for i in range(time_min, time_max, tww):
    count += 1
    ie = i + tww
    table_count.set_value(count, 'time_start', i)     # >= time_start
    table_count.set_value(count, 'time_end', ie)   # < time_end
    # ab
    temp = delay_ab[(delay_ab.index >= i) & (delay_ab.index < ie)]
    table_count.set_value(count, 'ab_count', len(temp))
    if len(temp) > 0:
        table_count.set_value(count, 'ab_mean', temp.values.mean())
        last_valid_value_ab = temp.values.mean()
    else:
        table_count.set_value(count, 'ab_mean', last_valid_value_ab)
    # ac
    temp = delay_ac[(delay_ac.index >= i) & (delay_ac.index < ie)]
    table_count.set_value(count, 'ac_count', len(temp))
    if len(temp) > 0:
        table_count.set_value(count, 'ac_mean', temp.values.mean())
        last_valid_value_ac = temp.values.mean()
    else:
        table_count.set_value(count, 'ac_mean', last_valid_value_ac)
    # au
    temp = delay_au[(delay_au.index >= i) & (delay_au.index < ie)]
    table_count.set_value(count, 'au_count', len(temp))
    if len(temp) > 0:
        table_count.set_value(count, 'au_mean', temp.values.mean())
        last_valid_value_au = temp.values.mean()
    else:
        table_count.set_value(count, 'au_mean', last_valid_value_au)
    # an
    temp = delay_an[(delay_an.index >= i) & (delay_an.index < ie)]
    table_count.set_value(count, 'an_count', len(temp))
    if len(temp) > 0:
        table_count.set_value(count, 'an_mean', temp.values.mean())
        last_valid_value_an = temp.values.mean()
    else:
        table_count.set_value(count, 'an_mean', last_valid_value_an)
    # ag
    temp = delay_ag[(delay_ag.index >= i) & (delay_ag.index < ie)]
    table_count.set_value(count, 'ag_count', len(temp))
    if len(temp) > 0:
        table_count.set_value(count, 'ag_mean', temp.values.mean())
        last_valid_value_ag = temp.values.mean()
    else:
        table_count.set_value(count, 'ag_mean', last_valid_value_ag)
```

```python
# Try machine learning, more specifically one class SVM, using sliding window to train the model

training_width = 24 * 10    # 10 days
trw = training_width

data_pd = table_count[['ab_mean', 'ac_mean', 'au_mean', 'an_mean', 'ag_mean']]

set_draw = set()

list_all = list()

for i in range(1 + trw, data_pd.index.max()):
    data_training = data_pd[(data_pd.index >= (i - trw)) & (data_pd.index < i)]
    clf = OneClassSVM(nu=0.261, gamma=0.05)
    clf.fit(data_training)
    data_new_line = data_pd[data_pd.index == i]
    result = clf.decision_function(data_new_line)
    y_pred = result.ravel()
    y_pred = y_pred[0]
    list_all.append(y_pred)

print(list_all)
```
