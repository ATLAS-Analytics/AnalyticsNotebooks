```python
#These lines set up inline plotting, and apply a standard size
%matplotlib inline
import matplotlib
matplotlib.rc('font', **{'size': 12})
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
def index(i,j):
    if i==0: return 0
    if i==1 and j==0: return 0
    if i==1 and j==1: return 1
    if i==2 and j==0: return 1
    if i==2 and j==1: return 0
    if i==3: return 1    
```

```python
def makePlots(data):
    activities = ["AOD","SUSY5"]
    variables = ["I/O intensity","Event rate","IObytesReadRate","IObytesWriteRate"]
    labels = ["I/O Intensity, MB/second",
              "Event processing rate, events/second",
              "Reading rate, MB/second",
              "Writing rate, MB/second"]
    limits = [50.0,100.0,50.0,20.0]
    num_bins = 100
    fig, axarr = plt.subplots(2,2,figsize=(15,15))
    i=0
    for variable,label,upperLimit in zip(variables,labels,limits):
        for activity in activities:
            x = np.asarray(data[(activity,variable)])
            axarr[index(i,0), index(i,1)].hist(x, num_bins, alpha=0.5, label=activity,range=[0.0, upperLimit])
            #axarr[index(i,0), index(i,1)].set_yscale("log")
            axarr[index(i,0), index(i,1)].set_xlabel(label)
            axarr[index(i,0), index(i,1)].legend(loc='upper right')
        i+=1
    fig.text(0.04, 0.5, 'Jobs', va='center', rotation='vertical')
    plt.show()
    return
```

```python
# Read in saved file
#data = pickle.load(open("io_test.p","rb"))
data = pickle.load(open("io_analysis.p","rb"))
```

```python
makePlots(data)
```
