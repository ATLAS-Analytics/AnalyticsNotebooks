```python
import pandas as pd
import numpy as np
import time
import os
from multiprocessing import Pool

opendataset = pd.read_json
                                                                                                                                                                        
basedir = 'windata/'

def analyse_link(filename):
    print 'Processing ', filename, '...'
    s = time.time()
    src, dst = filename.split('--')[1].split('__')
    df = opendataset(basedir + filename)
    d = {}
    d['%s:%s' % (src, dst)] = {}
    for k in df[df.keys()[0]][0].keys():
        #d['%s:%s' % (src, dst)][int(k)] = {}
        thr = np.array([x[k] for x in df[df.keys()[0]]]).transpose()[0]
        cls = np.array([x[k] for x in df[df.keys()[0]]]).transpose()[1]
        d['%s:%s' % (src, dst)][int(k)] = (np.std(thr), np.max(cls) - np.min(cls))
    #print 'Link %s:%s done in %.02f sec.' % (src, dst, (time.time() - s))
    return d


if __name__ == '__main__':
    ts = time.time()
    basedir = 'windata/'
    #l = os.listdir(basedir)
    l = filter(lambda x: x.endswith('.json'), os.listdir(basedir))
    l.sort()
    p = Pool(24)
    #window size
    s = time.time()
    #print "Calculating closenesses for windowsize =", win, 'hours'
    analy = p.map(analyse_link, l)
    print 'Done in %.02f min.' % ((time.time() - s)/60.)
    print 'Writing file...'
    s = time.time()
    analysis = {}
    for site in analy:
        analysis[site.keys()[0]] = site[site.keys()[0]]
    hugedf = pd.DataFrame(analysis)
    hugedf.to_hdf('allwins.hdf', 'w')
    print 'Done in %.02f seg.' % ((time.time() - s))
    print 'Total elapsed time %.02f min.' % ((time.time() - ts)/60.)
```
