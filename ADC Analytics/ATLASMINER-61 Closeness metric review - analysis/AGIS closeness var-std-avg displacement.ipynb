New AGIS closeness analysis

```python
%matplotlib inline
import matplotlib
matplotlib.rc('font', **{'size': 22})
```

```python
import json
import numpy as np
j = json.loads(open('closeness_var-std.json').read())
std = [0]*(len(j.keys())+1)
var = [0]*(len(j.keys())+1)
for i in j:                                             
    std[int(i.split('-')[1])] = j[i][1]
    var[int(i.split('-')[1])] = j[i][0]
j = json.loads(open('closeness_displacements.json').read())
cdeltas = [0]*(len(j.keys())+1)
errors = [0]*(len(j.keys())+1)
for i in j:
    cdeltas[int(i.split('-')[1])] = j[i][0]
    errors[int(i.split('-')[1])] = j[i][1]
```

```python
import matplotlib.pyplot as plt
plt.plot(std[1:], 'g.', label='Summed Standar Deviation')
plt.plot(var[1:], 'b.', label='Summed Variance')
plt.legend()
plt.title('AGIS Closeness Summed Variance for a given Window Size')
plt.xlabel('Window Size interval in hours')
plt.ylabel('Variance/Standar Deviation')
fig = plt.gcf()
fig.set_size_inches(17, 8)
```

```python
import matplotlib.pyplot as plt
plt.errorbar(x=np.array(range(1, len(cdeltas))),y=cdeltas[1:],yerr=errors[1:],\
             fmt='k.', ecolor='g', label='Avg closeness displacement 95% confidence.')
plt.legend()
plt.title('Average closeness displacement for a give Window Size')
plt.xlabel('Window Size interval in hours')
plt.ylabel('Avg. Closeness')
fig = plt.gcf()
fig.set_size_inches(17, 8)
```
