```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import pickle
plt.style.use('seaborn-paper')
%matplotlib inline
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
df=pd.read_pickle('august25_2.pkl')
```

```python
length=df.shape[0]
print (length)
df['rate']=0
for i in range(length):
    df['rate'].iloc[i]= 2045132800/df['download_time'].iloc[i]/1000
    
#df=df.drop_duplicates(subset='start1', keep='last')
df=df.sort_values(by=['cache','start1'])
df = df[df['cache'] != '']
length=df.shape[0]
print (length)
df
```

```python
def download_rates(cache, yLim):
    columns=list(df.columns.values)
    df_cache=pd.DataFrame(columns=columns)
    for i in range(length):
        if df['cache'][i]==cache or df['cache'][i]=='root://stash.osgconnect.net':
            df_cache=df_cache.append(df.iloc[[i]])
    hist, bins = np.histogram(df_cache.rate.values, bins=250)
    width = 0.7 * (bins[1] - bins[0])
    center = (bins[:-1] + bins[1:]) / 2
    plt.bar(center, hist, align='center', width=width*1.25)
    plt.axis([1, 130, 0, yLim])
    plt.xlabel('Download rate [MB/s]')
    plt.ylabel('Count')
    plt.show()

#download_rates('root://hcc-stash.unl.edu', 1000)
```

```python
download_rates('root://xrd-cache-1.t2.ucsd.edu', 1500)
```
