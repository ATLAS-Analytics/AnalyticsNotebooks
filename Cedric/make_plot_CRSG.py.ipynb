```python
%matplotlib inline
import matplotlib.pyplot as plt


TTC, RAW, AOD, ESD, DESD, DRAW, DAOD = [], [], [], [], [], [], []
first_row = 1
with open('backlog_T0_Export_new.txt', 'r') as f:
    for line in f:
        line = line.strip('\n')
        if first_row:
            first_row = 0
            print line.split()
        else:
            ttc, raw, aod, esd, desd, daod, draw = line.split()
            TTC.append(int(ttc))
            RAW.append(int(raw))
            AOD.append(int(aod))
            ESD.append(int(esd))
            DESD.append(int(desd))
            DRAW.append(int(draw))
            DAOD.append(int(daod))
print TTC, RAW, AOD, ESD, DESD, DRAW, DAOD
```

```python
plt.figure(figsize=(20,10))
p1 = plt.bar(TTC, [x/1000000. for x in RAW], color = 'red', label='RAW')
p2 = plt.bar(TTC, [x/1000000. for x in AOD], color = 'blue', label='AOD')
p3 = plt.bar(TTC, [x/1000000. for x in DESD], color = 'green', label='DESD')
p4 = plt.bar(TTC, [x/1000000. for x in ESD], color = 'orange', label='ESD')
p5 = plt.bar(TTC, [x/1000000. for x in DRAW], color = 'yellow', label='DRAW')
plt.legend(prop={'size':20})
plt.xticks(TTC, ["0", "1", "2", "3", "4", "5", ">5"], size=20)
plt.xlabel('NB days to complete', size=20)
plt.yticks(size=20)
plt.ylabel('Million files', size=20)
plt.show()
```
