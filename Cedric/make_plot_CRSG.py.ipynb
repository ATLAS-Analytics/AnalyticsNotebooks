```python
%matplotlib inline
import matplotlib.pyplot as plt


TTC, RAW, AOD, ESD, DESD, DAOD, DRAW = [], [], [], [], [], [], []
input_list = []
first_row = 1
with open('backlog_T0_Export_new.txt', 'r') as f:
    for line in f:
        line = line.strip('\n')
        if first_row:
            first_row = 0
            print line.split()
        else:
            input_list.append([int(x) for x in line.split()])
TTC, RAW, AOD, ESD, DESD, DAOD, DRAW = map(list, zip(*input_list))
print TTC, RAW, AOD, ESD, DESD, DAOD, DRAW
print "TTC RAW AOD ESD DESD DAOD DRAW"
for t in TTC:
    print TTC[t], RAW[t], AOD[t], ESD[t], DESD[t], DAOD[t], DRAW[t]
```

```python
# This plot is the backlog of T0 export in linear scale
plt.figure(figsize=(20,10))
p1 = plt.bar(TTC, [x/1000000. for x in RAW], color = 'red', label='RAW', align="center")
p2 = plt.bar(TTC, [x/1000000. for x in AOD], color = 'blue', label='AOD', align="center")
p3 = plt.bar(TTC, [x/1000000. for x in DESD], color = 'green', label='DESD', align="center")
p4 = plt.bar(TTC, [x/1000000. for x in ESD], color = 'orange', label='ESD', align="center")
p5 = plt.bar(TTC, [x/1000000. for x in DRAW], color = 'yellow', label='DRAW', align="center")
plt.legend(prop={'size':20})
plt.xticks(TTC, ["0", "1", "2", "3", "4", "5", ">5"], size=20)
plt.xlabel('NB days to complete', size=25)
plt.yticks(size=20)
plt.ylabel('Million files', size=25)
plt.show()
```

```python
# This plot is the backlog of T0 export in log scale
plt.figure(figsize=(20,10))
p1 = plt.bar(TTC, RAW, color = 'red', label='RAW', align="center")
p2 = plt.bar(TTC, AOD, color = 'blue', label='AOD', align="center")
p3 = plt.bar(TTC, DESD, color = 'green', label='DESD', align="center")
p4 = plt.bar(TTC, ESD, color = 'orange', label='ESD', align="center")
p5 = plt.bar(TTC, DRAW, color = 'yellow', label='DRAW', align="center")
plt.yscale('log') 
plt.legend(prop={'size':20})
plt.xticks(TTC, ["0", "1", "2", "3", "4", "5", ">5"], size=20)
plt.xlabel('NB days to complete', size=20)
plt.yticks(size=20)
plt.ylabel('NB files', size=20)
plt.show()
```
