# Checks for and removes eventual duplicated entries in JobsArchive index

```python
import collections
from elasticsearch import Elasticsearch
from elasticsearch.helpers import scan
from elasticsearch.helpers import bulk
from time import time

es = Elasticsearch(['atlas-kibana.mwt2.org:9200'],timeout=60)
```

```python
my_index = "jobs_archive_2016-09-*"

my_query = {
    "fields" : ["pandaid"],
    "query" : { "match_all": {} }
}
scroll = scan(es, query=my_query, index=my_index, scroll='5m', timeout="5m", size=10000)

count = 0

pids={}
bulk_body = []

for res in scroll:
#     if count==0:
#         print(res)
    if not count%100000:  print(count)
    count += 1
    
    pid=res['fields']['pandaid'][0]
    if pid in pids: 
        bulk_body.append('{"delete": {"_id": "'+res['_id']+'", "_index":"'+ res['_index'] +'", "_type":"'+ res['_type'] +'"}}')
    else:
        pids[pid]=1
        if len(bulk_body)>1000:
            es.bulk('\n'.join(bulk_body), request_timeout=1200)
            bulk_body=[]
            print('deleted 1k')
            
print ('deleting the rest')
if len(bulk_body)>0:
    print('deleting ', len(bulk_body))
    es.bulk('\n'.join(bulk_body), request_timeout=1200)
                
print ('total events:',count)
```
